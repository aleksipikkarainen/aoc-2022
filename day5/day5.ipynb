```python
import re

#A matrix which fits the supply stacks
def create2Darray(filename):
    height = 0
    width = 0
    f = open(filename, "r")
    while True:
        s = f.readline()
        height += 1
        if s[1].isdigit():
            width = len(s)/4
            break
    arr = [ [0]*(int(width)) for i in range(int(height-1))]
    f.close()
    return arr

#Adding supply stacks to matrix.
def fill2Darray(array, filename):
    f = open(filename, "r")
    counter = 0
    while True:
        s = f.readline()
        if s[1].isdigit():
            break
        else:
            for x in range(1,len(s),4):
                if s[x].isalpha():
                    array[counter][int(x/4)] = s[x]
            counter += 1
    f.close()
    return arr

def mover(array, order):
    command = re.findall(r"\d+",order)
    command = [eval(i) for i in command]
    print(command)
    try:
        for x in range(command[0]):
            array[command[2]-1].append(array[command[1]-1].pop())
    except:
        x = 1
    return array

#Transposes the supply stack matrix into neat stacks.
def organizeArray(arr):
    arr = [[row[i] for row in arr] for i in range(len(arr)+1)]
    for x in arr:
        x = x.reverse()
    for x in arr:
        while x[len(x)-1] == 0:
            x.pop()
    return arr

def generatePassword(arr):
    passw = ""
    for x in arr:
        if x[len(x)-1].isdigit():
            passw += x[len(x)-1]
    print(passw)
    
def debugPrinter(arr):
    for x in arr:
        print(x)
    print("-"*100)

arr = create2Darray("day5input.txt")
arr = fill2Darray(arr, "day5input.txt")
arr = organizeArray(arr)
print(arr)

f = open("day5input.txt", "r")

#Quick manual skip to the moving stacks -part.
for x in range(10):
    f.readline()
    
#Spaghetti loop for the moving stacks -part.
x = True
while x:
    s = f.readline()
    arr = mover(arr, s)
    print(s)
    debugPrinter(arr)
    if s == "":
        x = False
        
print(arr)  
```
