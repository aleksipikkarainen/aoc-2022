```python
def sameChar(str1, str2):
    for x in str1:
        for y in str2:
            if x == y:
                return x
    
def sameTriple(lines):
    for x in lines[0]:
        for y in lines[1]:
            if x == y:
                for z in lines[2]:
                    if x == z:
                        return x
def getValue(c):
    if c.islower():
        return ord(c)-96
    else:
        return ord(c)-38

f = open("day3input.txt", "r")
part1 = 0
for x in f:
    x.rstrip()
    first_half  = x[:len(x)//2]
    second_half = x[len(x)//2:]
    part1 += getValue(sameChar(first_half, second_half))
f.close()

f2 = open("day3input.txt", "r")
part2 = 0
lines = []
for y in f2:
    y = y.rstrip()
    if len(lines) < 3:
        lines.append(y)
    else:
        part2 += getValue(sameTriple(lines))
        print(lines)
        lines = []
        lines.append(y)
part2 += getValue(sameTriple(lines))
    
print(part1)
print(part2)
```
